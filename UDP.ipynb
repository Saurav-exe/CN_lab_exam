UDP Client–Server to Say “Hello”

```python
#udp_hello_server.py
import socket

HOST = "0.0.0.0"
PORT = 5000

server = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
server.bind((HOST, PORT))

print("UDP Server waiting for message...")

msg, addr = server.recvfrom(1024)
print("Client:", msg.decode())

server.sendto("Hello from Server!".encode(), addr)
```

```python
#udp_hello_client.py
import socket

HOST = "127.0.0.1"
PORT = 5000

client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

client.sendto("Hello from Client!".encode(), (HOST, PORT))

msg, addr = client.recvfrom(1024)
print("Server:", msg.decode())
```

CHAT using UDP

```python
#udp_p2p_chat.py
import socket
import threading

def listen(sock):
    while True:
        msg, addr = sock.recvfrom(1024)
        print("\nPeer:", msg.decode())

def main():
    local_port = int(input("Enter your listening port: "))
    remote_ip = input("Enter peer IP: ")
    remote_port = int(input("Enter peer port: "))

    sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    sock.bind(("0.0.0.0", local_port))

    threading.Thread(target=listen, args=(sock,), daemon=True).start()

    print("Chat started. Type messages and press Enter.")
    while True:
        message = input()
        sock.sendto(message.encode(), (remote_ip, remote_port))

if __name__ == "__main__":
    main()
```
